Pandas - DataFrame and Series


widely used for data analysis and data cleaning . It provides two data structure : Series and DataFrame . 
series - 1D array-like object
DataFrame - 2D , tabular data structure


```python
import pandas as pd
```

```python
## series
## one dimensional array-like object 
## hold any data type
import pandas as pd

data = [1,2,3,4,5]
series = pd.Series(data)
print("series :\n",series)
print(type(series))
```

```python
## create a seires from dictionary
data = {"a":1,"b":2,"c":1}
series_dic = pd.Series(data)
print(series_dic)
```

```python
data = [10,20,30]
index = ['a','b','c']
pd.Series(data,index = index)
```

```python
## data frame
## create a data frame from dictionary of list
data ={
    'Name': ["krish","john","jack"],
    'Age' : [25,30,45],
    'City':["Banglore","New York","Texas"],


}

df = pd.DataFrame(data)
df

```

```python
import numpy as np
np.array(df)
```
