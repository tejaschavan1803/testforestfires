```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan],
    'B': [4, np.nan, 6],
    'C': [7, 8, 9]
})


nv_filled = df.fillna(0)
print(nv_filled)
df
```

```python
## how to fill values in missing value
df_filled = df.fillna(0)
```

```python
## filling missing values with mean of the value
df['Sales_fillNA'] = df['sales'].fillna(df['sales'].mean())
df
```

```python
df.dtypes
```

```python
## renaming columns
df =df.rename(columns={'date':'Sales Date'})
df.head(5)
```

```python
## change the data type
df['Value_new'] = df['value'].astype(int)
df.head(5)
```

```python
## apply function
df['New Value'] =df['value'].apply(lambda x:x *2)
df.head(5)
```

```python
## data aggregating and grouping
grouped_mean = df.groupby("product")['value'].mean()
grouped_mean
```

```python
gruoped_sum = df.groupby(['product','region'])['value'].sum()
print(gruoped_sum)
```

```python
## mean
gruoped_mean = df.groupby(['product','region'])['value'].mean()
print(grouped_mean)
```

```python
## aggregate multiple function
group_agg = df.groupby('region')['value'].agg(['mean','sum','count'])
group_agg
```

```python
## mergine and joining dataframes
df1 = pd.DataFrame({'key':['A','B','C'],  'Value1' : [1,2,3]})
df2 = pd.DataFrame({'key':['A','B','D'],  'Value2' : [4,5,6]})
```

```python
df2
```

```python
## Merge dataframe on the 'Key columns'
pd.merge(df1,df2,on='key',how ='inner')
```

```python
## outer join
pd.merge(df1,df2,on='key',how ='outer')
```

```python
## left outer join -- left dataframe have more priority 
pd.merge(df1,df2,on='key',how ='left')
```

```python
## right outer join
pd.merge(df1,df2,on='key',how ='right')
```
